```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

csa_data = pd.read_csv('cybersecurity_attacks.csv')

attack_type = csa_data['Attack Type'].value_counts()
print(attack_type)

totval = len(csa_data)
print(totval)

ddos = 13428
malware = 13307
intrusion = 13265

ddos_freq = ddos/totval*100
malware_freq = malware/totval*100
intrusion_freq = intrusion/totval*100

at_in = attack_type.index
at_val = attack_type.values

pie_colors = ['FireBrick', 'DarkSlateBlue', 'DarkGreen']

fig = plt.figure(figsize=(10, 5))
plt.pie(at_val, labels = at_in, colors = pie_colors)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


cc_data = pd.read_csv('cybercrime_data.csv')

df = pd.DataFrame(cc_data)

attack_types = df['attack_type']
sec_tools = df['security_tools_used']

freqtable = pd.crosstab(attack_types, sec_tools)

plt.figure(figsize=(8, 6))
sns.heatmap(freqtable, cmap = 'coolwarm')
plt.show()
```
